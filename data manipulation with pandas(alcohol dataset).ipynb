# Create a dataframe of ten rows, four columns with random values. Write a Pandas program to highlight dataframeâ€™s specific columns with different colors.

```python
def highlight_cols(s, cols):
    color = 'blue'
    if s.name in cols:
        return ['background-color: {}'.format(color)] * len(s)
    else:
        return [''] * len(s)
```

```python
df2.style.apply(highlight_cols, cols=['A', 'C'])
```
